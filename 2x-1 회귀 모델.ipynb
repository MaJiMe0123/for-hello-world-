뉴런(Unit)이 한개, 층(layer)이 1층인 코드로 x와 y의 관계를 찾는 예문
(2x-1에 대한 예측 모델) 입력값을 기반으로 출력값을 찾는 회귀 모델

```python
import tensorflow as tf # tensorflow 를 라이브러리에서 불러오고 tensorflow를 tf란 단어로 정의하여 사용
import numpy as np # numpy를 라이브러리에서 부러오고 np라는 단어로 정의하여 사용
from tensorflow.keras import Sequential #tensorflow의 keras(API)를의 Sequential 클래스 불러오기
from tensorflow.keras.layers import Dense #TensorFlow의 Keras 라이브러리에서 Dense 레이어를 가져와서 사용할 준비 Dense = 완전 연결층으로 각 노드가
#이전의 모든층과 연결되어 있는 형태
dense = Dense(units=1, input_shape=[1]) #dense.get_weight 메서드 사용을 위한 dense 정의
model=Sequential([dense]) #레이어를 순서대로 쌓는 모델이면서 각 노드는 이전의 모든 노드와 연결된 Dense로 구성하고
#출력 뉴런은 1개로 지정한다, 또한 입력의 형태는 1차원 벡터로 지정하고 이를 model로 정의
model.compile(optimizer='sgd', loss='mean_squared_error') #옵티마이저(최적화 알고리즘)와 손실함수(손실or오차를 기반한 모델 학습이 잘되는지의 판단 척도)
#sgd = 확률적 경사 하강법으로 손실을 최소화 하는 방식으로 접근, mean_squared_error = 실제값과 예측값의 차이를 제곱하여 평균을 구하는 손실 함수
xs = np.array([-1.0, 0.0, 1.0, 2.0, 3.0, 4.0], dtype=float) # xs라는 넘파이 배열 생성 및 데이터타입을 float로 지정
ys = np.array([-3.0, -1.0, 1.0, 3.0, 5.0, 7.0], dtype=float)# ys라는 넘파이 배열 생성 및 데이터타입을 float로 지정
model.fit(xs,ys, epochs=500) # 입력 xs를 가지고 ys를 예측하는 학습을 시작, 에포크(전체 순회 횟수)는 500회
print(model.predict(np.array([10.0]))) # 학습한 모델에 입력값을 10.0으로 해서 그에대한 예측을 구사, *원문의 일반배열은 오류를 일으켜 넘파이배열로 변환
```

```python
print("신경망이 학습한 것 : {}".format(dense.get_weights()))#신경망이 학습한 가중치(기울기), 절편 출력
#  Y = 2X - 1과 매우 비슷한 것을 알 수 있음
```
