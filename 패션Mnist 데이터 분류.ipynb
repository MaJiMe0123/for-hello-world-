패션 Mnist 데이터셋을 이용하여 "분류"하는 모델 생성 및 테스트

```python
import tensorflow as tf
data = tf.keras.datasets.fashion_mnist # 패션mnist 데이터셋을 data로 정의
(training_images, training_labels),(test_images, test_labels) = data.load_data()
#load_data는 두개의 튜플을 반환 함 훈련데이터 튜플, 테스트 데이터 튜플로 반환되며
#다시 각 입력과 레이블로 나뉘어 짐. 즉 data(패션Mnist)를 훈련, 테스트 데이터로 분리
training_images = training_images/255.0
test_images = test_images/255.0
# 훈련, 테스트 이미지를 0 ~ 1의 값으로 매핑하여 0 ~ 255의 범위를 정규화 하는 과정
model = tf.keras.models.Sequential( #순서대로 쌓는 층으로 구성
    [tf.keras.layers.Flatten(input_shape = (28, 28)), #Flatten층(입력층)의 입력은 28*28의 1D벡터로 지정
     tf.keras.layers.Dense(128, activation=tf.nn.relu), #은닉층의 경우 dense형태의 128개의 뉴런과 렐루 활성화 함수 지정
     tf.keras.layers.Dense(10, activation=tf.nn.softmax)] #마지막 출력층은 dense형태이며 10개의 뉴런으로 구성하고 활성화 함수는 소프트맥스로 지정
)
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy',#(최적화알고리즘)은 아담으로 하고 손실함수는 희소한범주형크로스엔트로피로 함
              metrics=['accuracy']) #분석(metrics)은 정확도로 표시
model.fit(training_images, training_labels, epochs=5)#훈련데이터와 훈련레이블로 에포크(전체순회)5회 훈련
```

```python
model.evaluate(test_images, test_labels) # 테스트 이미지와 테스트 레이블을 통한 모델 평가
```

```python
model.evaluate(test_images, test_labels) # 테스트 이미지와 테스트 레이블을 통한 모델 평가
classifications = model.predict(test_images) # 테스트 이미지를 입력으로 모델을 통한 예측 결과를 classifications로 정의
print(classifications[0])#테스트 이미지 0번의 예측값을 10개의 뉴런에서 예측한 결과 출력
print(test_labels[0]) # 첫번째 인덱스의 레이블 값 출력
```
